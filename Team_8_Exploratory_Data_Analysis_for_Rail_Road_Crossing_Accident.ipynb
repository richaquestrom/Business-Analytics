## Rail Road Accidents:
In the United States, there are currently approximately 200,000 railroad grade crossings. Many commuters drive across railway crossings on a daily basis, never thinking about the dangers they may be exposing themselves to. While a train collision with a vehicle may appear implausible at first, such collisions are rather common in the United States.

Statistics on railroad accidents and motor vehicle accidents:


1. A railroad grade crossing accident is 40 times more likely to kill motor 
vehicle drivers and passengers than a motor vehicle on motor vehicle accident.

2. Vehicles were involved in 2,025 collisions at railroad grade crossings in 2016.

3. In 2016, 265 persons were killed in railroad grade crossing accidents.

4. In 2016, 798 persons were injured in railroad grade crossing accidents.

5. In the United States, a train-on-car collision occurs every two/three hours on average.

#Data Prediction in State with largest number of accidents(Texas):
Our goal is to make a simple model that can predict whether an accident will take place or not based on the historical data from columns having predictive power.

Converting Categorical Data to Numerical Data:

```python
# taking columns which have predictive power
accidents_prediction = (df_rail_road_crossing_accident.groupby([ 'Train Speed','Time','Month','Roadway Condition','Crossing Illuminated','State Name','Report Year','Estimated Vehicle Speed','Number People On Train','Employees Killed For Reporting Railroad', 
                                                                'Visibility','Employees Injured For Reporting Railroad','Passengers Killed For Reporting Railroad','Passengers Injured For Reporting Railroad','Crossing Users Killed For Reporting Railroad',
                                                                'Crossing Users Injured For Reporting Railroad','Highway User Position Code','Total Killed Form 57','Total Injured Form 57'])
                                                             ["Accidents"].sum()).to_frame(name = 'Total_Accidents').sort_values(['Total_Accidents'],ascending=False).reset_index() 
                                                             
accidents_prediction.replace(np.nan,0, inplace = True)
# adding condition to predict Accidents only for Texas
final_dataset = accidents_prediction[accidents_prediction['State Name'] == 'TEXAS']
df_dummified = pd.get_dummies(final_dataset, drop_first=True)   # One-hot encoding
df_dummified.head(3)
```

```python
# exporting to a csv file                                                             
accidents_prediction.to_csv(r'/content/accidents_prediction1.csv', index=False)
```

Observation:

* November, 2012 saw the maximum number of railroad crossing accidents.
* Majority of the accidents take place at train speeds of 60 mph and people crossing at speeds of 5-25 mph. 
* Accidents take place mainly during peak visibility during the day and between 3pm-5pm.
* The road condition at the time of accidents was mostly dry,there was no ice or water logged. 
* The accidents were particularly fatal for crossing users and passengers in the trains in most of the cases.

```python
X_df = df_dummified.drop('Total_Accidents', axis=1) 
y_df = df_dummified['Total_Accidents']
```

```python
#checking shape
print("X_df:", X_df.shape)
print("y_df:", y_df.shape)
```

Supervised learning with Gaussian Model:
From X_df and y_df creating the train/test splits.

* Setting 50% of the data to test and the rest data to train.
* Naming the resulting objects: Xtrain, Xtest, ytrain, ytest.

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_df, y_df, test_size=0.5, random_state=833)
```

```python
print("Xtrain shape:", Xtrain.shape)
print("Xtest shape:", Xtest.shape)
```

```python
from sklearn.naive_bayes import GaussianNB # 1. choose model class
model = GaussianNB()                       # 2. instantiate model
model.fit(Xtrain, ytrain)                  # 3. fit model to data
y_model = model.predict(Xtest)             # 4. predict on new data
```

```python
print("Our model is", round(sum(ytest == y_model)/len(y_model),2), "accurate!")
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

##Is our model good enough?

If we had said there will be no accidents, what the accuracy would have been?

Calculating the accuracy for our simplistic model below:

Here 0 represents no accident, anything not equal to zero represents an accident.

Checking goodness of the model:

```python
sum(ytest == 0)/len(ytest)
```

So, after all our ML model is not that great.

```python
P = sum(ytest != 0)
P
```

How many true positives did the model generate? 


```python
TP = sum((ytest != 0) & (y_model != 0))
TP
```

True positive sensitivity:

```python
TP/P
```

```python
N = sum(ytest == 0)
N
```

True negative rate (or specificity):


```python
TN/N
```
